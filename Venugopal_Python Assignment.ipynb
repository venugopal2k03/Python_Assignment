```python
'''
Q-8:Write a function type_counter that takes a list of different data types and returns a dictionary showing the count of each data type in the list.
'''
#declaring the function type_counter
def type_counter(input_list):
    type_count = {}
    for item in input_list:
        for data_type in [int, str, float, list, tuple]:
            if isinstance(item, data_type):
              #determining the data type of items
                data_type_name = data_type.__name__
                #incrementing the values
                if data_type_name in type_count:
                    type_count[data_type_name] += 1
                else:
                    type_count[data_type_name] = 1
    #returning the result
    return type_count
input_list = [1, "hello", 3.14, [1, 2, 3], (4, 5, 6),1.414]
#function calling
result = type_counter(input_list)
print(result)
```
